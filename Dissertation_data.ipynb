```python
import geopandas as gpd
import requests
import gzip
import json
from shapely.geometry import shape
from io import BytesIO

url = "https://minedbuildings.z5.web.core.windows.net/global-buildings/2025-02-25/global-buildings.geojsonl/RegionName=UnitedArabEmirates/quadkey=123023130/part-00179-5cf70943-9c5f-4fc6-94fb-43ce5feefa56.c000.csv.gz"

response = requests.get(url)
if response.status_code == 200:
    with gzip.GzipFile(fileobj=BytesIO(response.content)) as f:
        features = []
        for line in f:
            if line.strip():
                feature = json.loads(line)
                geom = shape(feature["geometry"])
                props = feature["properties"]
                props["geometry"] = geom
                features.append(props)

    gdf = gpd.GeoDataFrame(features, geometry="geometry", crs="EPSG:4326")
    print(gdf.head())
else:
    print(f"Download failed: {response.status_code}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# ───────────────────────────────────────────────────────────────
# 1. Load and clean
# ───────────────────────────────────────────────────────────────
url = ("https://github.com/ClaudiaTurner/GEOG5160M-Dissertation/"
       "blob/main/Data/Draft_csv.csv?raw=true")
df = pd.read_csv(url, encoding="cp1252")[["Suitable Area m²", "Cost (AED) / kWh"]]
df = df.dropna().sort_values("Suitable Area m²")

# ───────────────────────────────────────────────────────────────
# 2. Build a 500 m² rolling-mean series for the clean trend line
# ───────────────────────────────────────────────────────────────
bin_size = 500
df["area_bin"] = (df["Suitable Area m²"] // bin_size) * bin_size  # 0, 500, 1000, …

trend = (df.groupby("area_bin", as_index=False)
           ["Cost (AED) / kWh"]
           .mean()
           .sort_values("area_bin"))

# ───────────────────────────────────────────────────────────────
# 3. Plot
# ───────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(12, 6))

# scatter layer (raw rooftops)
ax.scatter(
    df["Suitable Area m²"],
    df["Cost (AED) / kWh"],
    s=30, alpha=0.5, label="Individual rooftops"
)

# smooth line (500 m² rolling mean)
ax.plot(
    trend["area_bin"],
    trend["Cost (AED) / kWh"],
    marker="o",
    linewidth=2.0,
    color="black",
    label=f"Mean every {bin_size} m²"
)

# — cosmetics —
ax.set_xlabel("Suitable Area (m²)")
ax.set_ylabel("Cost per kWh (AED)")
ax.set_title("Cost per kWh vs. Rooftop Area\n(scatter + rolling mean)")
ax.grid(True, linestyle="--", linewidth=0.4, alpha=0.7)
ax.legend()

# tidy x-axis
ax.xaxis.set_major_locator(mticker.MultipleLocator(1000))          # tick every 1 000 m²
ax.xaxis.set_major_formatter(mticker.StrMethodFormatter('{x:,.0f}'))
ax.margins(x=0.05)                                                 # 5 % breathing room

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# ───────────────────────────────────────────────────────────────
# 1 · Load & tidy data
# ───────────────────────────────────────────────────────────────
url = ("https://github.com/ClaudiaTurner/GEOG5160M-Dissertation/"
       "blob/main/Data/Draft_csv.csv?raw=true")

df = (pd.read_csv(url, encoding="cp1252")
        [["Suitable Area m²", "Cost (AED) / kWh"]]
        .dropna()
        .sort_values("Suitable Area m²"))

x = df["Suitable Area m²"].values          # strictly positive → OK for log scale
y = df["Cost (AED) / kWh"].values

# ───────────────────────────────────────────────────────────────
# 2 · Optional linear trend (computed in log space so it’s still linear)
# ───────────────────────────────────────────────────────────────
logx = np.log10(x)
m, b = np.polyfit(logx, y, 1)              # y = m·log10(x) + b
x_trend = np.geomspace(x.min(), x.max(), 300)
y_trend = m * np.log10(x_trend) + b

# ───────────────────────────────────────────────────────────────
# 3 · Plot
# ───────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(15, 6))
plt.style.use('default')                   # clean white background

# line through every point
ax.plot(x, y, linewidth=1.4, color="#c44e52", label="Connecting line")

# scatter points
ax.scatter(x, y, s=60, color="#c44e52", zorder=3, label="Rooftops")

# linear trend in log-x space
ax.plot(x_trend, y_trend, "--", linewidth=2.0, color="#4c72b0",
        label=f"Log-linear trend (slope ≈ {m:.4f})")

# ─ axis cosmetics ──────────────────────────────────────────────
ax.set_xscale('log')

# custom “nice” ticks: 1×10ⁿ and 2×10ⁿ & 5×10ⁿ up to the max area
nice = []
for exp in range(int(np.floor(np.log10(x.min()))),
                 int(np.ceil(np.log10(x.max()))) + 1):
    nice.extend([1, 2, 5])
    nice = [n * (10 ** exp) for n in [1, 2, 5]]
xticks = [tick for tick in
          np.concatenate([[1, 2, 5] * 10])  # build plenty, keep those ≤ max
          if x.min() * 0.9 <= tick <= x.max() * 1.1]
xticks = sorted(set(xticks))
ax.set_xticks(xticks)
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"{v:,.0f}"))

plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# y-axis formatting
ax.yaxis.set_major_formatter(mticker.StrMethodFormatter('{x:,.2f}'))

# light dotted major grid
ax.grid(True, which='major', linestyle=':', linewidth=0.6, alpha=0.8)

# labels, title, legend
ax.set_xlabel("Suitable Area (m²)  [log scale]")
ax.set_ylabel("Cost per kWh (AED)")
ax.set_title("Cost per kWh vs. Rooftop Area  (log-scaled x-axis, full data)")
ax.legend()

fig.tight_layout()
plt.show()
```
